```python
from avatar.language import WranglingLanguage
import avatar.settings
```

```python
language = WranglingLanguage()
```

```python
titanic2 = language.expand(titanic)
titanic2
```

```python
from avatar.filter import *

pruning = StackedFilter([MissingFilter(),
                         ConstantFilter(),
                         IdenticalFilter()])

preselection = StackedFilter([BijectiveFilter(),
                              UniqueFilter(),
                              CorrelationFilter()])

titanic2 = pruning.select(titanic2, target="Survived")
titanic2 = preselection.select(titanic2, target="Survived")
titanic2
```

```python
titanic3 = language.expand(titanic2)
titanic3
```

```python
features = [
    "Sex",
    "Split( )(Name)_1",
    "Age",
    "Pclass",
    "Fare",
    "OneHot()(Pclass)_3",
    "OneHot()(ExtractWord([Dr, Master, Miss, Mr, Mrs, Rev])(Name)_0)_Master"
]
```

```python
program = language.make_program(titanic, features)
program(titanic)
```
