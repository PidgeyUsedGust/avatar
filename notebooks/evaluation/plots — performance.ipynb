```python
import pathlib
import pandas as pd
import altair as alt
```

```python
def read_results(path):
    path = pathlib.Path(path)
    dataframes = list()
    for file in path.glob("*.csv"):
        if file.name.startswith("."):
            continue
        dataframes.append(pd.read_csv(file, index_col=0))
    data = pd.concat(dataframes, ignore_index=True)
    data = data.rename(columns={"i": "wrangling iteration"})
    return split_results(data)


def split_results(df):
    df["feature iterations"] = df["run"].str.extract(r"iterations=(\d+)")
    df["feature depth"] = df["run"].str.extract(r"depth=(\d+)")
    df["feature samples"] = df["run"].str.extract(r"sample=(\d+)")
    return df
```
