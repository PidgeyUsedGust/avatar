## string

String to string transformations.

```python
from avatar.transformations.string import *
```

```python
arguments = Split.arguments(nba.BIRTH_DATE)

# run and show
new = nba.BIRTH_DATE.to_frame()
for argument in arguments:
    new = pd.concat((new, Split(*argument)(nba.BIRTH_DATE)), axis=1)

display(new)
```

```python
from avatar.transformations.string import SplitAlign
```

```python
SplitAlign.arguments(titanic.Name)
```

```python
c = nba.BIRTH_DATE

display(SplitAlign.arguments(c))

for argument in SplitAlign.arguments(c):
    sa = SplitAlign(*argument)
    display(sa(c))
    display(sa.call2(c))
```

### ExtractNumber

Extract numbers from text.

```python
from avatar.transformations.string import ExtractNumber


test_column = automobile[1]

# get argumnets
arguments = ExtractNumber.arguments(test_column)
print(arguments)

# run and show
new = test_column.to_frame()
for argument in arguments:
    new[str(argument)] = ExtractNumber(*argument)(test_column)

display(new)
```

```python
ExtractNumber.arguments(titanic.Cabin)
```
