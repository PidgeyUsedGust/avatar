```python
import json
import pandas as pd
```

```python
df = pd.read_csv("data/android_2.csv")
df.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from avatar.supervised import *
from avatar.settings import Settings

Settings.verbose = True
```

```python
game = Game(estimator=DecisionTreeRegressor(max_depth=4))
pool = Pool(AverageSkill)
tournament = Tournament(game, pool, n_games=200, team_size=32, exploration=0.2)
tournament.initialise(df, target="logprice")
```

```python
tournament.play()
```

```python
with open("data/android_2_ranking.json", "w") as f:
    json.dump(tournament.pool.distribution, f)
```

```python
agame = Game(estimator=DecisionTreeRegressor(max_depth=4))
apool = Pool(AverageSkill)
atournament = AnnealingTournament(agame, apool, exploration=0.2)
atournament.initialise(df, target="logprice")
```

```python
atournament.play()
```

```python
with open("data/android_2_ranking_cooldown.json", "w") as f:
    json.dump(atournament.pool.distribution, f, indent=2)
```
