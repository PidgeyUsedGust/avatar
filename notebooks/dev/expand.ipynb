```python
from avatar.language import WranglingLanguage
from avatar.filter import *
from avatar.expand import Tracker, Expander, _hash
from avatar.utilities import make_lower
```

```python
android = make_lower(pd.read_csv("../../data/raw/demo/android.csv"))
android = android.drop('price', axis=1)
android.head()
```

```python
expander = Expander(WranglingLanguage())
expanded = expander.expand(android, exclude=["logprice"])
```

```python
expanded2 = expander.expand(expanded)
```

```python
expanded2.head()
```

```python
preselector = StackedFilter([
    UniqueFilter(),
    MutualInformationFilter()
])
preselected = preselector.select(expanded2)
```

```python
preselected.head()
```

```python
expanded2.to_csv("data/android_2.csv", index=None)
expander.tracker.to_file("data/android_2_tracking.json")
```
