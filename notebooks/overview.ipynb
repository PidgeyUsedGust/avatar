## wrangle

Create a language with default transformations.

```python
from tqdm.notebook import tqdm
from avatar.language import WranglingLanguage

language = WranglingLanguage()
transformations = language.transformations(titanic)
len(transformations)
```

Create new dataset by applying all transformations.

```python
titanic_new = titanic
for transformation in tqdm(transformations):
    titanic_new = transformation(titanic_new)
titanic_new.head()
```
